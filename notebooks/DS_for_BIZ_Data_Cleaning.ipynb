```python
import pandas as pd
import os

# The path to the dataset files is stored in the 'path' variable from the previous cell
dataset_path = path

# List files in the dataset directory
files_in_dataset = os.listdir(dataset_path)
print("Files in the dataset directory:", files_in_dataset)

# Iterate through each file in the directory
for file_name in files_in_dataset:
    # Check if the file is a CSV file
    if file_name.endswith('.csv'):
        file_path = os.path.join(dataset_path, file_name)
        print(f"\nLoading file: {file_name}")
        try:
            # Load the CSV file into a DataFrame
            df = pd.read_csv(file_path)
            print(f"DataFrame '{file_name}' loaded successfully:")
            display(df.head())
        except Exception as e:
            print(f"An error occurred while loading '{file_name}': {e}")
```

```python
# Load the olist_orders_dataset.csv file
orders_df = pd.read_csv(os.path.join(dataset_path, 'olist_orders_dataset.csv'))

# Check for missing values in the orders DataFrame
print("Missing values in the orders DataFrame:")
print(orders_df.isnull().sum())
```

```python
# Drop rows with missing values in 'order_delivered_customer_date'
orders_df_cleaned = orders_df.dropna(subset=['order_delivered_customer_date'])

# Verify that the rows have been dropped
print("Shape of the original DataFrame:", orders_df.shape)
print("Shape of the cleaned DataFrame:", orders_df_cleaned.shape)
```

# Task
Visualize the percentage of late orders in the train and test sets as a function of the split date, where the split date is the breakpoint for dividing the dataset.

## Define a range of split dates

### Subtask:
Define a range of potential dates to split the dataset.


**Reasoning**:
I need to determine the range of dates in the `orders_df_fully_cleaned` DataFrame to define appropriate split points.



```python
min_date = orders_df_fully_cleaned['order_purchase_timestamp'].min()
max_date = orders_df_fully_cleaned['order_purchase_timestamp'].max()

print(f"Minimum order purchase timestamp: {min_date}")
print(f"Maximum order purchase timestamp: {max_date}")
```

**Reasoning**:
Based on the minimum and maximum dates, I will generate a sequence of monthly dates to use as potential split points for the dataset and store them in a variable.



**Reasoning**:
The previous code failed because a `Timedelta` object does not have an attribute `n`. I need to calculate the number of months between the start and end dates correctly to generate the sequence of monthly split points.



```python
# Define the start and end dates for generating split points
start_date = min_date.to_period('M').to_timestamp()
end_date = max_date.to_period('M').to_timestamp()

# Calculate the number of months between the start and end dates
num_months = (end_date.to_period('M') - start_date.to_period('M')).n

# Generate a list of monthly split dates
split_dates = [start_date + pd.offsets.MonthBegin(i) for i in range(num_months + 1)]

print("Generated split dates:")
print(split_dates)
```

## Iterate and calculate late percentages

### Subtask:
For each split date, split the data, calculate the percentage of late orders in the train and test sets.


**Reasoning**:
Iterate through the split dates, split the data, and calculate the percentage of late orders in both the train and test sets.



```python
train_late_percentages = []
test_late_percentages = []

for split_date in split_dates:
    train_df_iter = orders_df_fully_cleaned[orders_df_fully_cleaned['order_purchase_timestamp'] < split_date].copy()
    test_df_iter = orders_df_fully_cleaned[orders_df_fully_cleaned['order_purchase_timestamp'] >= split_date].copy()

    # Calculate percentage of late orders in train set
    if not train_df_iter.empty:
        train_late_percentage = (train_df_iter['isLate'].sum() / len(train_df_iter)) * 100
    else:
        train_late_percentage = 0.0
    train_late_percentages.append(train_late_percentage)

    # Calculate percentage of late orders in test set
    if not test_df_iter.empty:
        test_late_percentage = (test_df_iter['isLate'].sum() / len(test_df_iter)) * 100
    else:
        test_late_percentage = 0.0
    test_late_percentages.append(test_late_percentage)

print("Train late percentages:", train_late_percentages)
print("Test late percentages:", test_late_percentages)
```

## Store the results

### Subtask:
Store the split dates and the corresponding late percentages for both train and test sets.


**Reasoning**:
Create a DataFrame to store the split dates and late percentages and display the head of the DataFrame.



```python
# Create a DataFrame to store the results
late_percentage_df = pd.DataFrame({
    'split_date': split_dates,
    'train_late_percentage': train_late_percentages,
    'test_late_percentage': test_late_percentages
})

# Display the first few rows of the DataFrame
display(late_percentage_df.head())
```

## Visualize the trend

### Subtask:
Plot the late percentages against the split dates to show the trend.


**Reasoning**:
Plot the late percentages against the split dates to visualize the trend for both train and test sets.



```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='split_date', y='train_late_percentage', data=late_percentage_df, label='Train Late Percentage')
sns.lineplot(x='split_date', y='test_late_percentage', data=late_percentage_df, label='Test Late Percentage')
plt.title('Late Delivery Percentage Trend by Split Date')
plt.xlabel('Split Date')
plt.ylabel('Late Delivery Percentage')
plt.legend()
plt.show()
```
